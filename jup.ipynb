## 2. Using Jupyter Notebook

There are many IDEs/ code editors for Python...

But the most commonly used IDE for Machine Learning/ Data Science is Jupyter Notebook.

<b> Jupyter Notebook </b> uses sequences of cells to make coding easier. It mixes code and natural language text, making it reader-friendly.

### 2.1 Some shortcuts in Using Jupyter Notebook

#### Navigation:

``` Up Arrow Key ``` : Move to previous cell 

``` Down Arrow Key ```: Move to next cell

#### Editing and Executing Cells:

```Enter```: Edit a Cell  (Alternative: Double click the cell)

```Esc```: Exit Editing Mode (Alternative: Click another cell)

```Ctrl + Enter```: Execute the cell and exit editing mode

#### Adding Cells

```A```: add a cell Above

```B```: add a cell Below

#### Deleting Cells

```D,D```: Delete selected cell

#### Other Editing Tools

```C```: Copy cell

```X```: Cut cell

```V```: Paste cell

```Z```: Undo delete cell

```S```: Save


```H```: See all shortcuts

### 2.2 Markdown Cells
